```python
# Data Manipulation
import pandas as pd
import numpy as np

# Data Visualization
import matplotlib.pyplot as plt
from matplotlib import rcParams # used to customize the apprence of your plot
import seaborn as sns

# applying styling
plt.style.use('ggplot')
rcParams['figure.figsize'] = (12, 6)

# using Sklearn to import a dataset
from sklearn.datasets import load_wine
```

```python
# Loading wine dataset
wine = load_wine()

# converting the dataset to pandas dataframe
df = pd.DataFrame(data = wine.data, columns = wine.feature_names)
df
```

```python
# creating target column
df['target'] = wine.target
```

```python
df.target.value_counts()
```

```python
# Plotting a data using matplotlib library
df.target.value_counts().plot(kind='bar')
plt.title('Value count of Target variable')
plt.xlabel('wine type')
plt.ylabel('count')
plt.show()
```

```python
df.magnesium.describe()
```

```python
df.magnesium.hist()
plt.title('Distribution of Magnesium')
```

```python
# Evaluating kurtosis and asymmetry

print(f"Skewness: {df['magnesium'].skew()}")
print(f"Kurtosis: {df['magnesium'].kurt()}")
```

```python
# Doing it another way
import scipy.stats as stats
skewness = stats.skew(df['magnesium'])
kurtosis = stats.kurtosis(df['magnesium'])
print(skewness)
print(kurtosis)
```

```python
# exploring relationship between variables
# Seaborn pairplot (scatter plot)
sns.pairplot(df)
```

```python
# Creating box plot using seaborn library
sns.catplot(x = 'target', y = 'alcohol', data = df, kind = 'box', aspect=1.5)
plt.title('Boxplot for target vs alcohol')
plt.show()
```

```python
sns.catplot(x='target', y='flavanoids', data=df, kind='box', aspect=1.5)
plt.title('Boxplot for target vs flavanoids')
plt.show()
```

```python
sns.catplot(x='target', y='color_intensity', data=df, kind='box', aspect=1.5)
plt.title('Boxplot for target vs color_intensity')
plt.show()
```

```python
sns.catplot(x='target', y='proline', data=df, kind='box', aspect=1.5)
plt.title('Boxplot for target vs proline')
plt.show()
```

```python
# creating scatter plot to visualize the points belinging to different wine class
sns.scatterplot(x='proline', y='flavanoids', hue='target', data=df, palette='Dark2', s=80)
plt.title('Relationship between prolin, flavanoidsand target')
plt.show()
```

```python
# ScatterPlot
sns.scatterplot(x='alcohol', y='color_intensity', hue='target', data=df, palette='Dark2', s=80)
plt.title('Relationship Between alcohol, color_intensity and target')
plt.show()
```
