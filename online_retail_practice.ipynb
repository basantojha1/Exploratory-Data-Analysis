# Analysis

```python
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
# Unique Customer ID in the Dataset
df['CustomerID'].nunique()
```

```python
df['CustomerID'].value_counts(normalize=True)
```

```python
# Plotting the customer ID in the chart to check which customer has appeared most frequently for purchase.
df['CustomerID'].value_counts(normalize=True)[:5].plot.bar()
plt.title("Most Purchase by Customer ID")
plt.show()
```

```python
# Taking Out date and time column seperately from the InvoiceDate Column
df['Date'] =df['InvoiceDate'].dt.date
df['Time'] =df['InvoiceDate'].dt.time
```

```python
df.head()
```

```python
# Droping the column InvoiceDate
df.drop('InvoiceDate', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Converting Date and Time to string
df['Date'] = df['Date'].astype(object)
df['Time'] = df['Time'].astype(object)
```

```python
# Extracting Year Month and Day from Date
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
```

```python
# using Seaborn library for distribution plot
plt.figure(figsize =(10,6))
sns.countplot(x = df['Month'])
plt.title('Sales by Month')
```

# The total sales of the product has increased in the recent Month.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Quantity', data=df, estimator=sum, ci=None, palette='viridis')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
